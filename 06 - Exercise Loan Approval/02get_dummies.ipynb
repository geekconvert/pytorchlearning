```python
import pandas as pd

df = pd.DataFrame({
    "loan_intent":["PERSONAL", "EDUCATION"], 
    "person_income": [30000, 50000]
})
df = pd.get_dummies(df, columns=["loan_intent"])
df.head()
```

```python
df.values
```

```python
import torch

# this will fail because df.values has dtype object. The reason is that one single numpy array needs to be the same type for all things 
X = torch.tensor(df.values, dtype=torch.float32)
print(X)
```

```python
df = pd.DataFrame({
    "loan_intent":["PERSONAL", "EDUCATION"], 
    "person_income": [30000, 50000]
})
df = pd.get_dummies(df, columns=["loan_intent"]).astype(dtype="float32")
print(df.values)

#this will work now
X = torch.tensor(df.values, dtype=torch.float32)
print(X)
```
