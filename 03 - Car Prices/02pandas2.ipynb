```python
import pandas as pd

df = pd.read_csv("./data/used_cars.csv")
df.head()
```

```python
price = df["price"]
price = price.str.replace("$", "")
price = price.str.replace(",", "")
price = price.astype(int)

df['price'] = price #price column replaced
df.head()

```

```python
df['price'].mean()
```

```python
print(df['brand'] == 'Hyundai') # true false series for Hyundai brand

df_hyundai = df[df["brand"] == "Hyundai"] # only Hyundai rows 
print(df_hyundai["price"].mean())

df_ford = df[df['brand'] == 'Ford'] # only Ford rows 
print(df_ford['price'].mean())
```
