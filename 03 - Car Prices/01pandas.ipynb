```python
price = df["price"] # when we are trying to analyze the data is that sometimes the data is not in the form that we need. For example, this price column here is not in the form that we can easily work with.
price = price.str.replace("$", "")
price = price.str.replace(",", "")
price = price.astype(int)
price.hist()
print(price < 300000) # So here we are creating this series with true false values that just define whether they, um, whether this condition matches or not.
print(price[price < 300000]) #  all the values are taken, where there was a true here, but values are going to be ignored if there was a false here at this place.
# price[price < 300000].hist()

# price = price.str.replace("$", "")
# price = price.str.replace(",", "")
# price = price.astype(int)
# print(price[price < 300000])
# price[price < 300000].hist()
```
