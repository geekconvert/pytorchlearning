```python
import pandas as pd
#  usually a CSV file in the first row it already, or it just contains the names of the columns that we got here, our file immediately starts with the data.
#by providing names parameter pandas will no longer treat the first line of the file as Names or labels for the columns, but it will use whatever provided here.
# And in that case, um, pandas will no longer treat the first line of the file as Names or labels for the columns, but it will use whatever I provide here.
df = pd.read_csv('data/SMSSpamCollection', sep='\t', names=['type', 'message'])
df.head()
df['spam'] = df['type'] == "spam"
df.head()
```

```python
# by default, a lot of these operations on the data frame, they always create a copy. So a copy has been created.
# inplace  means that we do not want a copy to be created. Everything should be done in place, meaning, that the column will be dropped and the data frame up here will be modified by this operation.
# or we can do like this also df = df.drop('type', axis=1)
df.drop('type', axis=1, inplace=True)
df.head()
```

```python
# If we want to train a neuron here, it's best if our type here would not be like a string ham or spam, but just a number or a true false value.
print("SPAM messages:")
print(len(df[df['spam'] == True]))
```

```python
print("Not SPAM messages:")
print(len(df[df["spam"] == False]))
```
